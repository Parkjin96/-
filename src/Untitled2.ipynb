```python
import pandas
import time
from pandas import DataFrame as df
import datetime
import folium
from math import sin, cos, sqrt, atan2, radians
from math import pow, degrees, radians, atan2 
from scipy import cos, sin, arctan, sqrt, arctan2 
map_osm = folium.Map(location=[37.245049,127.079311], zoom_start=17)
want=[]
csv_df = pandas.read_csv('4_1_pjw.csv',
                         sep=',',
                         header=0, 
                         parse_dates=True,
                         encoding = 'ISO-8859-1')

del csv_df["version"]
del csv_df["creator"]
del csv_df["ns1:name"]
del csv_df["ns1:desc"]
for i in range(len(csv_df)-1):
    
    csv_df["ns1:time"][i]
    time = csv_df["ns1:time"][i].replace("Z","")
    time2 = time.replace("T","")
    time3 = time2.replace(time2[18:22],"")
    time4 = csv_df["ns1:time"][i+1].replace("Z","")
    time5 = time4.replace("T","")
    time6 = time5.replace(time5[18:22],"")
    


    time8=datetime.datetime.strptime(time3,"%Y-%m-%d%H:%M:%S")

    time9=datetime.datetime.strptime(time6,"%Y-%m-%d%H:%M:%S")
    
    timet = time9-time8
    time10 = datetime.datetime.strptime("00:00","%M:%S")
    time11 = datetime.datetime.strptime("00:08","%M:%S")
    time12 = datetime.datetime.strptime("00:10","%M:%S")
    timez=time11-time10
    timeq=time12-time10
    
    ele = csv_df["ns1:ele"][i+1] - csv_df["ns1:ele"][i]
    lat = csv_df["lat"][i+1] - csv_df["lat"][i]
    lon = csv_df["lon"][i+1] - csv_df["lon"][i]
    

    R = 6373.0

    lat1 = radians(csv_df["lat"][i+1])
    lon1 = radians(csv_df["lon"][i+1])
    lat2 = radians(csv_df["lat"][i])
    lon2 = radians(csv_df["lon"][i])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    sec = timet.seconds
    v = distance/sec
   
        
    if timez <= timet <=timeq and (abs(ele)<=1.2) and (abs(lon) <= 0.00012) and (abs(lat) <= 0.00012)and(0.0012<=v<=0.0019):
        want.append(i)

   

for j in range(len(want)-1):
    folium.Marker([csv_df["lat"][want[j]],csv_df["lon"][want[j]]]).add_to(map_osm)


map_osm
```
