```python
import pandas
import time
from pandas import DataFrame as df
import datetime
import folium
from math import sin, cos, sqrt, atan2, radians
from math import pow, degrees, radians, atan2 
from scipy import cos, sin, arctan, sqrt, arctan2
map_osm = folium.Map(location=[37.245049,127.079311], zoom_start=17)
want=[]
csv_df = pandas.read_csv('3_28_pjw.csv',
                         sep=',',
                         header=0, 
                         parse_dates=True,
                         encoding = 'ISO-8859-1')
aa = 0.0015
bb = 0.003
del csv_df["version"]
del csv_df["creator"]
del csv_df["ns1:name"]
del csv_df["ns1:desc"]
for i in range(len(csv_df)-3):
    
    csv_df["ns1:time"][i]
    time = csv_df["ns1:time"][i].replace("Z","")
    time2 = time.replace("T","")
    time3 = time2.replace(time2[18:22],"")
    time4 = csv_df["ns1:time"][i+1].replace("Z","")
    time5 = time4.replace("T","")
    time6 = time5.replace(time5[18:22],"")
    


    time8=datetime.datetime.strptime(time3,"%Y-%m-%d%H:%M:%S")

    time9=datetime.datetime.strptime(time6,"%Y-%m-%d%H:%M:%S")
    
    timet = time9-time8
    time10 = datetime.datetime.strptime("00:00","%M:%S")
    time11 = datetime.datetime.strptime("00:06","%M:%S")
    time12 = datetime.datetime.strptime("00:30","%M:%S")
    timez=time11-time10
    timeq=time12-time10
    
    ele = csv_df["ns1:ele"][i+1] - csv_df["ns1:ele"][i]
    lat = csv_df["lat"][i+1] - csv_df["lat"][i]
    lon = csv_df["lon"][i+1] - csv_df["lon"][i]
    

    R = 6373.0

    lat1 = radians(csv_df["lat"][i+1])
    lon1 = radians(csv_df["lon"][i+1])
    lat2 = radians(csv_df["lat"][i])
    lon2 = radians(csv_df["lon"][i])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    sec = timet.seconds
    v = distance/sec
   
        
    lat1 = csv_df["lat"][i+1]
    lat2 = csv_df["lat"][i]
    long1 = csv_df["lon"][i+1]
    long2 = csv_df["lon"][i]
    Lat1= radians(lat1)
    Lat2 = radians(lat2)
    Long1 = radians(long1)
    Long2 = radians(long2)
    y = sin(Long2-Long1)*cos(Lat2)
    x = cos(Lat1)*sin(Lat2) - sin(Lat1)*cos(Lat2)*cos(Long2-Long1)
    #print(degrees(atan2(y, x)))
    #print("----------------------")
    lat3 = csv_df["lat"][i+2]
    lat4 = csv_df["lat"][i+1]
    long3 = csv_df["lon"][i+2]
    long4 = csv_df["lon"][i+1]
    Lat3= radians(lat3)
    Lat4 = radians(lat4)
    Long3 = radians(long3)
    Long4 = radians(long4)
    z = sin(Long4-Long3)*cos(Lat4)
    q = cos(Lat3)*sin(Lat4) - sin(Lat3)*cos(Lat4)*cos(Long4-Long3)
    #print(degrees(atan2(z, i)))
    #print(";;;;;;;;;;;;;;;;;")
    lat5 = csv_df["lat"][i+3]
    lat6 = csv_df["lat"][i+2]
    long5 = csv_df["lon"][i+3]
    long6 = csv_df["lon"][i+2]
    Lat5= radians(lat5)
    Lat6 = radians(lat6)
    Long5 = radians(long5)
    Long6 = radians(long6)
    m = sin(Long6-Long5)*cos(Lat6)
    p = cos(Lat5)*sin(Lat6) - sin(Lat5)*cos(Lat6)*cos(Long6-Long5)

    deg1 = degrees(atan2(z,q))
    deg2 = degrees(atan2(y,x))
    deg3 = degrees(atan2(m,p))
    d_m1 = deg2-deg1
    d_m2 = deg3-deg2
    d_m3 = deg3-deg1
    
    
    
    
    
    if  aa<= v and v <= bb:
        want.append(distance)

   

for j in range(len(want)-3):
    folium.Marker([csv_df["lat"][want[j]],csv_df["lon"][want[j]]]).add_to(map_osm)


map_osm
```
